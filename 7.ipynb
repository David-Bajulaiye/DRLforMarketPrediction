```python
## Close Prices data frame

# Reset the Index to tic and date
df_prices = data_df.reset_index().set_index(['tick', 'date']).sort_index()

# Get all the Close Prices
df_close = pd.DataFrame()

for ticker in filtered_stocks:
    # Extract the close prices series for each ticker
    series = df_prices.xs(ticker).close
    df_close[ticker] = series

# # Concatenate all series at once to create the df_close DataFrame
# df_close = pd.concat(data_dict, axis=1)
```

```python
# Check the first few rows of the date column to confirm the format
print(data_df['date'].head())  # Replace 'date' with the actual column name if different

# If the date format is inconsistent, you can try to infer it
data_df['date'] = pd.to_datetime(data_df['date'], errors='coerce', dayfirst=True)

# Alternatively, if you know the correct format, you can specify it like so:
# data_df['date'] = pd.to_datetime(data_df['date'], format='%d/%m/%Y', errors='coerce')

# After cleaning, check for NaT values in the date column
print(data_df[data_df['date'].isna()])

# Now that the date column is cleaned and consistently formatted, you can continue with your previous operations
df_close.index = pd.to_datetime(df_close.index, errors='coerce', dayfirst=True).tz_localize(None)
df_close_full_stocks.index = pd.to_datetime(df_close_full_stocks.index, errors='coerce', dayfirst=True).tz_localize(None)
data_df.index = pd.to_datetime(data_df.index, errors='coerce', dayfirst=True).tz_localize(None)

# Proceed with your splitting logic as before
train_pct = 0.8  # percentage of train data
date_list = list(data_df.date.unique())  # List of dates in the data

date_list_len = len(date_list)  # len of the date list
train_data_len = int(train_pct * date_list_len)  # length of the train data

train_start_date = date_list[0]
train_end_date = date_list[train_data_len - 1]  # Corrected to ensure the end date is within the train set

test_start_date = date_list[train_data_len]  # This should start right after the train data
test_end_date = date_list[-1]

# Split the Close Prices data
prices_train_data = df_close.loc[df_close.index <= train_end_date]
prices_test_data = df_close.loc[df_close.index >= test_start_date]

# Split the Close Prices of all stocks
prices_full_train = df_close_full_stocks.loc[df_close_full_stocks.index <= train_end_date]
prices_full_test = df_close_full_stocks.loc[df_close_full_stocks.index >= test_start_date]

# Split the whole dataset
train_data = data_df.loc[data_df.index <= train_end_date]
test_data = data_df.loc[data_df.index >= test_start_date]

# Print the shapes of the datasets to confirm the split
print(f"Train Data Shape: {train_data.shape}")
print(f"Test Data Shape: {test_data.shape}")
print(f"Closeprices_Train Data Shape: {prices_train_data.shape}")
print(f"Closeprices_Test Data Shape: {prices_test_data.shape}")
print(f"Closeprices of all stocks_train Data Shape: {prices_full_train.shape}")
print(f"Close prices of all stocks_ Test Data Shape: {prices_full_test.shape}")
```
